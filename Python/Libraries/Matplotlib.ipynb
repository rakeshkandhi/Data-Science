# plt.plot()

```python
x = np.array(np.linspace(10, 20, 5, dtype=int))
y = np.array(np.linspace(1, 10, 5, dtype=int))
```

```python
plt.figure(
    facecolor="#e6e6e6", edgecolor="#111"
)  # setting background color of the figure

# Plotting the first set of data
plt.plot(
    x,  # x-coordinates of the first set of data points
    y,  # y-coordinates of the first set of data points
    marker=".",  # marker style for the first set of data points
    linestyle="-",  # style of line connecting the first set of data points
    color="orange",  # color of the line for the first set of data points
    mec="red",  # marker edge color for the first set of data points
    ms=12,  # marker size for the first set of data points
    mfc="pink",  # marker face color for the first set of data points
    linewidth=2,  # width of the line for the first set of data points
    label="hello",
)
mat = [[11, 13, 15, 17, 19], [1, 3, 4, 5, 7]]
plt.plot(
    mat[0],
    mat[1],
    marker="D",
    linestyle="-",
    color="crimson",
    mec="orange",
    ms=6,
    mfc="gold",
    linewidth=2,
    label="Random plot",
)
plt.xlabel("X-Axis")  # labeling the x-axis
plt.ylabel("Y-Axis")  # labeling the y-axis
plt.title("Simple line chart")  # setting the title of the plot
plt.legend(loc="upper left", bbox_to_anchor=[1, 1])  # adding a legend
plt.grid(color="#8c8c8c", linestyle="--", linewidth=0.5)  # adding grid lines
ax = plt.gca()
ax.set_facecolor("#f2f2f2")
plt.show()  # displaying the plot
```

```python
plt.style.use("ggplot")
plt.plot(x, y)
plt.plot(mat[0], mat[1])
plt.show()
```
